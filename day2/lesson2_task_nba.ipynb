```python

rows_before = len(df)

df_clean = df.copy()

# убираем строки где points пустые
df_clean = df_clean.dropna(subset=['points'])

# убираем строки где numMinutes = 0 или пустые
df_clean = df_clean.dropna(subset=['numMinutes'])
df_clean = df_clean[df_clean['numMinutes'] != 0]

# преобразуем дату и извлекаем год
df_clean['gameDateTimeEst'] = pd.to_datetime(df_clean['gameDateTimeEst'])
df_clean['Season_Year'] = df_clean['gameDateTimeEst'].dt.year

rows_after = len(df_clean)
lost = rows_before - rows_after
pct_lost = lost / rows_before * 100

print(f'Строк до очистки:  {rows_before}')
print(f'Строк после:       {rows_after}')
print(f'Потеряно:          {lost} ({pct_lost:.1f}%)')
```

```python
df_clean['Decade'] = (df_clean['Season_Year'] // 10 * 10).astype(str) + 'е'

decade_stats = (
    df_clean
    .groupby('Decade', sort=True)
    .agg(
        Avg_Points   =('points',        'mean'),
        Avg_Assists  =('assists',        'mean'),
        Avg_Rebounds =('reboundsTotal',  'mean'),
        Count        =('points',         'count'),
    )
    .round(2)
    .reset_index()
)

decade_stats
#Больше всего очков игроки набирали в 1960-е — в среднем 12.4 очка за игру, это пик за всю историю. После этого результативность постепенно снижалась вплоть до 2010-х годов, где  
  #опустилась до минимума — около 9.6 очка. Самый резкий обрыв произошёл на переходе от 1960-х к 1970-м: тогда НБА резко расширилась — появилось много новых команд, и места в       
  #составах заняли менее результативные игроки, что «разбавило» общую статистику. В 2020-х виден небольшой рост обратно до 10.5 — это эпоха «трёхочкового бума», когда вся лига      
  #начала активнее атаковать из-за дуги.
```

```python
df_clean['numMinutes'] = pd.to_numeric(df_clean['numMinutes'], errors='coerce')
df_clean['points']    = pd.to_numeric(df_clean['points'],    errors='coerce')

player_stats = (
    df_clean
    .groupby(['personId', 'firstName', 'lastName'], as_index=False)
    .agg(
        Total_Points  = ('points',     'sum'),
        Total_Minutes = ('numMinutes', 'sum'),
        Games         = ('points',     'count'),
    )
)
player_stats['Player'] = player_stats['firstName'] + ' ' + player_stats['lastName']

# Топ-20 по общим очкам
top20_total = (
    player_stats
    .nlargest(20, 'Total_Points')
    [['Player', 'Total_Points', 'Games']]
    .reset_index(drop=True)
)
top20_total.index += 1

# Топ-20 по очкам за 36 минут (минимум 100 игр)
player_stats['Pts_per_36'] = (player_stats['Total_Points'] / player_stats['Total_Minutes'] * 36).round(2)
top20_per36 = (
    player_stats[player_stats['Games'] >= 100]
    .nlargest(20, 'Pts_per_36')
    [['Player', 'Pts_per_36', 'Games']]
    .reset_index(drop=True)
)
top20_per36.index += 1

# Таблицы рядом
side_by_side = pd.concat(
    [top20_total.add_prefix('Total__'), top20_per36.add_prefix('Per36__')],
    axis=1
)
print('=== Топ-20: общие очки vs очки за 36 минут ===\n')
print(side_by_side.to_string())

# Сколько игроков совпадает
overlap = set(top20_total['Player']) & set(top20_per36['Player'])
print(f'\nСовпадений в обоих списках: {len(overlap)}')
if overlap:
    print('Игроки:', ', '.join(sorted(overlap)))
```

## Выводы

**Совпадений: 0 — почему так мало**

Два списка отбирают игроков из разных эпох. Топ по общим очкам — это звёзды с длинными карьерами (LeBron, Kareem, Kobe): они набрали много просто потому что играли 15–20 лет. Топ по Pts/36 оказался полностью забит игроками **1940-х годов** — и это сигнал о проблеме с данными: у многих из них минуты либо не записывались вообще (получается деление почти на ноль → `inf`), либо записаны некорректно. Поэтому пересечение нулевое — списки смотрят в разные стороны.

---

**Что такое Pts_per_36**

Это очки, которые игрок набрал бы за стандартные 36 минут игры — если бы играл с той же скоростью что и обычно. Метрика убирает несправедливость: резервный игрок выходит на 12 минут и набирает 8 очков — это лучше, чем кажется. `Pts_per_36 = Total_Points / Total_Minutes × 36`. В современной НБА это стандартный способ сравнивать игроков с разным игровым временем.

---

**Кто лучший игрок**

По общим очкам — бесспорно **LeBron James** (53 845 очков, 2037 игр). Но если смотреть на эффективность — **Michael Jordan**: 38 415 очков всего за 1258 игр, то есть ~30.5 очка за игру. Это лучшее соотношение среди игроков топ-20 по тотальным очкам в современной эре. Топ по Pts/36 для 1940-х доверять не стоит — там проблема с качеством данных о минутах.


```python
season_stats = (
    df_clean
    .groupby('Season_Year', as_index=False)
    .agg(
        Avg_Points   = ('points',    'mean'),
        Num_Players  = ('personId',  'nunique'),
    )
    .round({'Avg_Points': 2})
)

top10_seasons = (
    season_stats
    .nlargest(10, 'Avg_Points')
    .reset_index(drop=True)
)
top10_seasons.index += 1

print('=== Топ-10 самых результативных сезонов ===\n')
print(top10_seasons.to_string())

print('\n--- Все сезоны (по убыванию среднего) ---\n')
print(season_stats.sort_values('Avg_Points', ascending=False).to_string(index=False))
```

## Вывод

Все десять самых результативных сезонов в истории НБА приходятся на **1960-е годы** — лидер 1961 год с 12.49 очка в среднем за игру. Это эпоха до введения строгих правил защиты и активного расширения лиги: команд было мало, игроки звёздного уровня выходили на площадку почти без замен и набирали огромное количество очков. После 1970 года результативность начала падать по мере роста числа игроков и усиления защитной тактики — и больше к уровню 1960-х так и не вернулась.
