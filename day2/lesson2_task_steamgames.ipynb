# Steam Games Dataset — Исследование данных

```python
# 1. Убираем строки с пустым Reviews и Genres
df_clean = df[
    df['Reviews'].notna() & df['Reviews'].str.strip().ne('') &
    df['Genres'].notna()  & df['Genres'].str.strip().ne('')
].copy()

# 2. Price → число (на случай если есть знак $)
df_clean['Price'] = (
    df_clean['Price']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.strip()
)
df_clean['Price'] = pd.to_numeric(df_clean['Price'], errors='coerce')

# 3. Убираем строки где цена отрицательная или > 1000
df_clean = df_clean[(df_clean['Price'] >= 0) & (df_clean['Price'] <= 1000)]

# 4. Release date → datetime, новая колонка Release Year
df_clean['Release date'] = pd.to_datetime(df_clean['Release date'], format='mixed', errors='coerce')
df_clean['Release Year'] = df_clean['Release date'].dt.year
```

```python
total = len(df)
clean = len(df_clean)
lost  = total - clean

print(f"Строк в df:        {total}")
print(f"Строк в df_clean:  {clean}")
print(f"Потеряли:          {lost} строк ({lost / total * 100:.1f}%)")
df_clean.head()
```

90% игр в Steam — призраки без отзывов. Из 122,000 игр реально существуют только 12,000

```python
import numpy as np

# Рейтинг = % положительных отзывов (Steam-стиль)
total_rev = df_clean['Positive'] + df_clean['Negative']
df_clean['Rating'] = (df_clean['Positive'] / total_rev.replace(0, np.nan) * 100).round(1)

# Ценовые категории
conditions = [
    df_clean['Price'] == 0,
    df_clean['Price'] <  5,
    df_clean['Price'] <= 20,
    df_clean['Price'] <= 40,
]
choices = [
    'Бесплатные (= $0)',
    'Дешёвые    (< $5)',
    'Средние    ($5–20)',
    'Дорогие    ($20–40)',
]
df_clean['Price Category'] = np.select(conditions, choices, default='Премиум     (> $40)')

# Сводная таблица в нужном порядке
order = [
    'Бесплатные (= $0)',
    'Дешёвые    (< $5)',
    'Средние    ($5–20)',
    'Дорогие    ($20–40)',
    'Премиум     (> $40)',
]

result = (
    df_clean
    .groupby('Price Category', sort=False)
    .agg(
        Количество_игр  = ('Name',   'count'),
        Средний_рейтинг = ('Rating', 'mean'),
    )
    .reindex(order)
    .round({'Средний_рейтинг': 1})
)

result
```

```python
# Разбиваем жанры и разворачиваем каждый в отдельную строку
genre_df = (
    df_clean[['Name', 'Genres', 'Rating']]
    .assign(Genres=df_clean['Genres'].str.split(','))
    .explode('Genres')
)
genre_df['Genres'] = genre_df['Genres'].str.strip()

# Топ-10 жанров по количеству игр
top10_genres = genre_df['Genres'].value_counts().head(10).index

# Таблица: только топ-10, отсортирована по рейтингу
result_genres = (
    genre_df[genre_df['Genres'].isin(top10_genres)]
    .groupby('Genres')
    .agg(
        Количество_игр  = ('Name',   'count'),
        Средний_рейтинг = ('Rating', 'mean'),
    )
    .round({'Средний_рейтинг': 1})
    .sort_values('Средний_рейтинг', ascending=False)
)

result_genres
```

```python
# Диагностика
print("Колонки df_clean:")
print(list(df_clean.columns))
print("\nRelease Year — топ значений:")
print(df_clean['Release Year'].value_counts().head(10))
print(f"\nNaN в Release Year: {df_clean['Release Year'].isna().sum()} из {len(df_clean)}")
```

```python
# В этом датасете даты хранятся в колонке 'Name' (сдвиг колонок из-за индекса CSV)
# 'Release date' содержит диапазоны владельцев ("0 - 20000"), а не даты
df_clean['Release Year'] = pd.to_datetime(
    df_clean['Name'], format='mixed', errors='coerce'
).dt.year

result_years = (
    df_clean[df_clean['Release Year'].between(2005, 2024)]
    .groupby('Release Year')
    .agg(
        Количество_игр  = ('AppID',  'count'),
        Средний_рейтинг = ('Rating', 'mean'),
    )
    .round({'Средний_рейтинг': 1})
    .sort_values('Средний_рейтинг', ascending=False)
)

display(result_years)
```

```python
df_clean['total_reviews'] = df_clean['Positive'] + df_clean['Negative']

top_games = (
    df_clean[
        (df_clean['Rating']        >= 90) &
        (df_clean['total_reviews'] >= 10_000)
    ]
    [['AppID', 'Rating', 'Positive', 'Negative', 'total_reviews', 'Required age', 'Genres', 'Release Year']]
    .rename(columns={'AppID': 'Name', 'Required age': 'Price'})
    .sort_values(['Rating', 'total_reviews'], ascending=False)
    .reset_index(drop=True)
)

print(f"Найдено игр: {len(top_games)}")
display(top_games)
```

```python
# Разбиваем жанры, добавляем флаг "бесплатная" (реальная цена — в Required age)
genre_free = (
    df_clean[['Genres', 'Required age']]
    .assign(
        Genres  = df_clean['Genres'].str.split(','),
        is_free = (df_clean['Required age'] == 0),
    )
    .explode('Genres')
)
genre_free['Genres'] = genre_free['Genres'].str.strip()

result_free = (
    genre_free
    .groupby('Genres')
    .agg(
        Всего_игр      = ('is_free', 'count'),
        Бесплатных     = ('is_free', 'sum'),
    )
    .query('Всего_игр >= 50')
    .assign(Процент_бесплатных=lambda x: (x['Бесплатных'] / x['Всего_игр'] * 100).round(1))
    .sort_values('Процент_бесплатных', ascending=False)
    .head(10)
)

display(result_free)
```

```python
# Взрываем жанры
genre_exp = (
    df_clean[['Genres', 'Rating', 'total_reviews']]
    .assign(Genres=df_clean['Genres'].str.split(','))
    .explode('Genres')
)
genre_exp['Genres'] = genre_exp['Genres'].str.strip()

# 1. Рейтинг — все игры
all_games = (
    genre_exp
    .groupby('Genres')
    .agg(
        Кол_все        = ('Rating', 'count'),
        Рейтинг_все    = ('Rating', 'mean'),
    )
    .round({'Рейтинг_все': 1})
)

# 2. Рейтинг — только игры с total_reviews >= 1000
filtered = (
    genre_exp[genre_exp['total_reviews'] >= 1000]
    .groupby('Genres')
    .agg(
        Кол_1000       = ('Rating', 'count'),
        Рейтинг_1000   = ('Rating', 'mean'),
    )
    .round({'Рейтинг_1000': 1})
)

# Объединяем рядом, убираем редкие жанры (< 20 игр)
comparison = (
    all_games
    .join(filtered, how='left')
    .query('Кол_все >= 20')
)

# Разница: положительная = стало лучше после фильтра, отрицательная = был завышен
comparison['Разница'] = (comparison['Рейтинг_1000'] - comparison['Рейтинг_все']).round(1)

comparison = comparison.sort_values('Разница')  # сначала самые завышенные (отрицательная разница)

print("Жанры отсортированы: сверху — где рейтинг БЕЗ фильтра был сильно завышен")
print("Разница = Рейтинг_1000 - Рейтинг_все  (отрицательная = ложный вывод)\n")
display(comparison)
```

 Кол_1000     │ игр с ≥ 1000 отзывов          │                                                                                          
  │ Рейтинг_1000 │ рейтинг только популярных игр

 Как найти ложный вывод:                                                                                                                                                           
  - Разница сильно отрицательная → жанр казался лучшим, но рейтинг был раздут малоизвестными играми с 5-10 отзывами                             
  - Разница сильно положительная → жанр недооценён: популярные игры в нём значительно лучше среднего

```python
result_devs = (
    df_clean
    .groupby('Developers')
    .agg(
        Количество_игр  = ('AppID',  'count'),
        Средний_рейтинг = ('Rating', 'mean'),
    )
    .query('Количество_игр >= 5')
    .round({'Средний_рейтинг': 1})
    .sort_values('Средний_рейтинг', ascending=False)
    .head(10)
)

display(result_devs)
```

# 🎮 Steam Games — Итоги расследования

## Данные
- Из **122 611** игр в Steam только **12 056 (9.8%)** имеют хоть один отзыв — остальные **90.2% существуют номинально**, без единого игрока

## Открытие 1 — Цена и качество
- Игры за **$5–20 имеют лучший рейтинг (87.5%)** — дороже бесплатных (79.3%) и дешевле премиума, но обгоняют и тех, и других; дорогие игры свыше $40 дают лишь 81.7%

## Открытие 2 — Жанры
- **Indie** — крупнейший жанр (9 165 игр), но лидер по рейтингу — **Casual (82.5%)**; аутсайдер топ-10 — **Free To Play (77.2%)** — самый распространённый и самый низкооценённый

## Открытие 3 — Лучшие годы
- Качество стабильно растёт: **2021–2024 — рейтинг 83–87%**, тогда как **2010 — исторический минимум (70.9%)** при 35 играх; 2024 уже имеет 1 072 игры с рейтингом 86.8%

## Открытие 4 — Бесплатные vs Платные
- **98% игр жанра Free To Play действительно бесплатны** (694 из 708); на втором месте Massively Multiplayer — **59.3%** бесплатных; все остальные жанры остаются преимущественно платными (9–14%)

## ⚠️ Пойманная ловушка
- **2005 год казался лучшим** по рейтингу (93.6%) — но там **ровно 1 игра**; в таблице жанров **Violent** имел рейтинг 68.3% без фильтра, а с фильтром ≥1000 отзывов остался **1 игра с рейтингом 61.4%** — разрыв −6.9 пунктов; большинство жанров наоборот растут с фильтром: **Early Access 77.8% → 84.0%** (+6.2) — малоизвестные провальные игры тянули весь жанр вниз
